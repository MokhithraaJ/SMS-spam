```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import SVC
```

```python
ds = pd.read_csv("/content/SMSspam.csv", encoding='latin1')
```

```python
ds
```

```python
ds.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'], axis =1 , inplace = True)
```

```python
ds
```

```python
mail_data = ds.where((pd.notnull(ds)),'')
```

```python
mail_data.isnull()
```

```python
from sklearn.preprocessing import LabelEncoder
label_encode = LabelEncoder()
label = label_encode.fit_transform(ds.Category)
ds['Category'] = label
```

```python
ds
```

```python
x = ds['Message']
```

```python
y= ds['Category']
```

```python
x
```

```python
y
```

```python
ds['Category'].value_counts()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x , y ,test_size =0.2,random_state = 30)
```

```python
x_train.shape
```

```python
feature_extraction = TfidfVectorizer(min_df = 1, stop_words = 'english', lowercase =True)
```

```python
x_train_features= feature_extraction.fit_transform(x_train)
x_test_features= feature_extraction.transform(x_test)
```

```python
y_train = y_train.astype('int')
y_test = y_test.astype('int')
```

```python
print(x_train_features)
```

```python
linear_model = LogisticRegression()
```

```python
linear_model.fit(x_train_features,y_train)
```

```python
y_predict = linear_model.predict(x_test_features)
```

```python
linear_model.score(x_test_features,y_test)
```

```python
nb_classifier = MultinomialNB()
```

```python
nb_classifier.fit(x_train_features,y_train)
```

```python
nb_classifier.score(x_test_features,y_test)
```

```python
linear_svm =SVC(kernel ='linear')
```

```python
linear_svm.fit(x_train_features,y_train)
```

```python
linear_svm.score(x_test_features,y_test)
```

```python
print("THE ACCURACY OF Naive Bayes is :",nb_classifier.score(x_test_features,y_test))
print("THE ACCURACY OF LOGISTIC is :",linear_model.score(x_test_features,y_test))
print("THE ACCURACY OF Support Vector Machines is :",linear_svm.score(x_test_features,y_test))
```

```python
input_mail = ["Ok lar... Joking wif u oni..."]
```

```python
input_data = feature_extraction.transform(input_mail)
```

```python
prediction = linear_svm.predict(input_data)
```
